```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'Resources/Medical_Conditions_update.csv'
```

```python
med_cond = pd.read_csv(file)
med_cond
```

```python
med_cond_df = pd.DataFrame(med_cond)
print(med_cond_df)
```

```python
x_value = med_cond_df['Medical_Condition']
y_value = med_cond_df['Children_Count']
z_value = med_cond_df['Adults_Count']
 
tick_locations = []
for x in x_value:
    tick_locations.append(x)
    
   
plt.figure(figsize=(10,10))
plt.barh(x_value, y_value, facecolor="red", alpha=0.75, align="center")
plt.xticks(tick_locations, x_value)

plt.xlabel('Medical Condition')
plt.ylabel('# of Deaths')
plt.title('Number of Deaths based on Medical Condition for Children')
plt.show()
 
```

```python
plt.figure(figsize=(10, 10))
plt.pie(med_cond_df['Children_Count'],
       labels=med_cond_df['Medical_Condition'],
       autopct='%1.1f%%' )
plt.title('Percentage of Medical Condtions')
plt.show()
```

```python
plt.plot(x_value, y_value, z_value)
plt.xlabel('Medical Condtions')
plt.ylabel('# of Deaths')
plt.title('# Number of Deaths based on Medical Condtions')
plt.show()
```

```python
med_cond_df.to_csv('Resources/med_cond_df.csv')
```
